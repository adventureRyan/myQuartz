```python
import pandas as pd
import matplotlib.pyplot as plt

# 读取 Excel 文件
file_path = 'average_precipitation_2022.xlsx'
df = pd.read_excel(file_path)

# 创建绘图窗口
plt.figure(figsize=(10, 8))

# 绘制经纬度的散点图，颜色代表降雨量
scatter = plt.scatter(df['Longitude'], df['Latitude'], c=df['Average_Precipitation'], cmap='viridis', s=10)

# 添加颜色条，表示降雨量
cbar = plt.colorbar(scatter)
cbar.set_label('Average Precipitation (mm)')

# 添加标题和标签
plt.title('Average Precipitation for 2022')
plt.xlabel('Longitude')
plt.ylabel('Latitude')

# 显示图像
plt.show()
```
